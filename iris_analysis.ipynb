Veri seti, 3 farklı iris türü için 50 örnek içerir: Iris setosa, Iris versicolor ve Iris virginica.
Veri setinde 4 adet özellik bulunur:

Sepal uzunluğu (cm)
Sepal genişliği (cm)
Petal uzunluğu (cm)
Petal genişliği (cm)

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the Iris data set
data = pd.read_csv('IRIS.csv')
df=data.copy()
```

# Visualizations

```python
sns.pairplot(df, hue='species', height=2.5)
plt.show()

sns.violinplot(x='species', y='sepal_length', data=df)
plt.show()

sns.violinplot(x='species', y='sepal_width', data=df)
plt.show()

sns.violinplot(x='species', y='petal_length', data=df)
plt.show()

sns.violinplot(x='species', y='petal_width', data=df)
plt.show()
```

```python
# Box plots to compare species for each feature
fig, axs = plt.subplots(ncols=4, figsize=(16,4))
for i, feature in enumerate(['sepal_length', 'sepal_width', 'petal_length', 'petal_width']):
    sns.boxplot(x='species', y=feature, data=df, ax=axs[i])
plt.show()
```

```python
# Swarm plots to compare species for each feature
fig, axs = plt.subplots(ncols=4, figsize=(16,4))
for i, feature in enumerate(['sepal_length', 'sepal_width', 'petal_length', 'petal_width']):
    sns.swarmplot(x='species', y=feature, data=df, ax=axs[i])
plt.show()
```

```python
# Scatter plot matrix to compare features with each other
sns.pairplot(df, hue='species', diag_kind='hist', height=2.5)
plt.show()
```

```python
# Strip plots to compare sepal length and petal length
sns.stripplot(x='species', y='sepal_length', data=df, jitter=True)
plt.show()
```

Petal uzunluğu ve petal genişliği özellikleri arasında yüksek bir pozitif ilişki olduğu gözlemlenir.
Yani, çiçeklerin petal uzunluğu arttıkça petal genişliği de artar. 
Ayrıca, setosa türü ile diğer iki tür arasında, diğer özelliklerdeki farklar kadar açık bir ayrım olmadığı da gözlemlenebilir. 
Bu nedenle, setosa türü genellikle diğer iki türden daha kolay ayırt edilir.
